# Pandas Profiling

Pandas profiling is an interactive report from pandas. The report contains missing values, duplicate values, skeweness, and correlation.

It is really usefull to show the first insight of our data. Pandas profiling also is appropriate for a large dataset

```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
directory = '/Users/ari/Documents/Data Ari/Data Science/Project/Ecommerce'
data_2009_2010 = pd.read_excel(f'{directory}/online_retail.xlsx', sheet_names = 'Year 2009-2010')
data_2010_2011 = pd.read_excel(f'{directory}/online_retail.xlsx', sheet_names = 'Year 2010-2011')
data = pd.concat([data_2009_2010, data_2010_2011])
```

```python
profile_report = data.profile_report(explorative=True, html={'style': {'full_width': True}})
profile_report
```
